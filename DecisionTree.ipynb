```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
df = pd.read_csv('processed.cleveland.data',header=None)
```

```python
df.columns = ['age',
              'sex',
              'cp',
              'restbp',
              'chol',
              'fbs',
              'restecg',
              'thalach',
              'exang',
              'oldpeak',
              'slope',
              'ca',
              'thal',
              'hd']
              
```

```python
df_no_missing = df.loc[(df['ca'] != '?')&(df['thal'] != '?')]
```

```python
len(df_no_missing)
```

```python
X = df_no_missing.drop('hd', axis=1).copy()
X.head()
```

```python
y = df_no_missing['hd'].copy()
y.head()
```

```python
X.dtypes
```

```python
X['cp'].unique()
```

```python
pd.get_dummies(X, columns=['cp']).head()
```

```python
X_encoded = pd.get_dummies(X, columns=['cp',
                                      'restecg',
                                      'slope',
                                      'thal'])
X_encoded.head()
```

```python
y.unique()
```

```python
y_not_zero_index = y > 0 # build index for the row that has heart-disease
y_not_zero_index
```

```python
y[y_not_zero_index] = 1
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, random_state=42)
clf_dt = DecisionTreeClassifier(random_state=42)
clf_dt = clf_dt.fit(X_train, y_train)
```

```python
plt.figure(figsize=(15, 7.5))
```

```python
plot_tree(clf_dt,
          filled=True,
          rounded=True,
          class_names=["No HD", "Yes HD"],
          feature_names=X_encoded.columns);
```

```python
predict_x = clf_dt.predict(X_test)
```

```python
predict_x
```

```python
actual = y_test.to_numpy()
```

```python
confusion_matrix = metrics.confusion_matrix(actual, predict_x)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

cm_display.plot()
plt.show()
```

```python
path = clf_dt.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas
ccp_alphas = ccp_alphas[:-1]

clf_dts = []

for ccp_alpha in ccp_alphas:
    clf_dt = DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    clf_dt.fit(X_train, y_train)
    clf_dts.append(clf_dt)
```

```python
train_scores = [clf_dt.score(X_train, y_train) for clf_dt in clf_dts]
test_scores = [clf_dt.score(X_test, y_test) for clf_dt in clf_dts]

fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alphas, train_scores, marker='o', label="train", drawstyle="steps-post")
ax.plot(ccp_alphas, test_scores, marker='o', label="test", drawstyle="steps-post")
ax.legend()
```

```python
clf_dt = DecisionTreeClassifier(random_state=42, ccp_alpha=0.016)
scores = cross_val_score(clf_dt, X_train, y_train, cv=5)
```

```python
df = pd.DataFrame(data={'tree': range(5), 'accuracy': scores})
```

```python
df.plot(x='tree', y='accuracy', marker='o', linestyle='--')
```

```python
alpha_loop_values = []
for ccp_alpha in ccp_alphas:
    clf_dt = DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    scores = cross_val_score(clf_dt, X_train, y_train, cv=5)
    alpha_loop_values.append([ccp_alpha, np.mean(scores), np.std(scores)])
    
alpha_results = pd.DataFrame(alpha_loop_values,
                             columns=['alpha','mean_accuracy','std'])

alpha_results.plot(x='alpha',
                   y='mean_accuracy',
                   yerr='std',
                   marker='o',
                   linestyle='--')
```

```python
alpha_results[(alpha_results['alpha'] > 0.014)
              &
              (alpha_results['alpha'] < 0.015)]
```

```python
ideal_ccp_alpha = alpha_results[(alpha_results['alpha'] > 0.014)
                                & 
                                (alpha_results['alpha'] < 0.015)]['alpha']
ideal_ccp_alpha
```

```python
ideal_ccp_alpha = float(ideal_ccp_alpha)
ideal_ccp_alpha
```

```python
clf_dt_pruned = DecisionTreeClassifier(random_state=42, ccp_alpha=ideal_ccp_alpha)
clf_dt_pruned = clf_dt_pruned.fit(X_train, y_train)
```

```python
predict_pruned = clf_dt_pruned.predict(X_train)
```

```python
actual = y_train.to_numpy()
```

```python
confusion_matrix = metrics.confusion_matrix(actual, predict_pruned)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

cm_display.plot()
plt.show()
```

```python
plt.figure(figsize=(15, 7.5))
plot_tree(clf_dt_pruned,
          filled=True,
          rounded=True,
          class_names=["No HD", "Yes HD"],
          feature_names=X_encoded.columns);
```
